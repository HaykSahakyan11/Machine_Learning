### P6. Write a Python code to output the following graph. (or a similar one). 
```
The productivity values for the 3 groups are the following: 

g1 = [80, 60, 80, 67, 34]
g2 = [45, 67, 23, 89, 10]
g3 = [30, 90, 50, 20, 90]
```

````python
g1 = [80, 60, 80, 67, 34]
g2 = [45, 67, 23, 89, 10]
g3 = [30, 90, 50, 20, 90]

index = np.arange(len(g1))
bar_width = 0.2

plt.bar(index - bar_width, g1, bar_width, color='blue', label='Group1')

plt.bar(index, g2, bar_width, color='orange', label='Group2')

plt.bar(index + bar_width, g2, bar_width, color='green', label='Group3')

plt.xlabel('Productivity')
plt.ylabel('Groups')
plt.legend(loc='upper right')
plt.title('Productivity during the week')
````
